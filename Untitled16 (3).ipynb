```python
def plot_graphs(groundtruth,prediction,title):        
    N = 9
    ind = np.arange(N)  # the x locations for the groups
    width = 0.27       # the width of the bars

    fig = plt.figure()
    fig.suptitle(title, fontsize=12)
    ax = fig.add_subplot(111)
    rects1 = ax.bar(ind, groundtruth, width, color='r')
    rects2 = ax.bar(ind+width, prediction, width, color='g')

    ax.set_ylabel("Amount of rainfall")
    ax.set_xticks(ind+width)
    ax.set_xticklabels( ('APR', 'MAY', 'JUN', 'JUL','AUG', 'SEP', 'OCT', 'NOV', 'DEC') )
    ax.legend( (rects1[0], rects2[0]), ('Ground truth', 'Prediction') )

#     autolabel(rects1)
    for rect in rects1:
        h = rect.get_height()
        ax.text(rect.get_x()+rect.get_width()/2., 1.05*h, '%d'%int(h),
                ha='center', va='bottom')
    for rect in rects2:
        h = rect.get_height()
        ax.text(rect.get_x()+rect.get_width()/2., 1.05*h, '%d'%int(h),
                ha='center', va='bottom')
#     autolabel(rects2)

    plt.show()
```

# Markov Chain Models: Predict rainfall by modeling the transition probabilities between different states (e.g., dry, light rain, heavy rain)
To illustrate this, let's assume we have historical data with the states ('Dry', 'Light Rain', 'Heavy Rain') recorded for each day. We will:

1-Calculate the transition probabilities from the historical data.
2-Use these probabilities to create a Markov Chain model.
3-Generate predictions for future rainfall states.

```python
import numpy as np
import pandas as pd

# Load your dataset
# Replace 'your_dataset.csv' with the path to your actual dataset
df = pd.read_csv('district_wise_rainfall_normal.csv')

# Define the thresholds for the states
dry_threshold = 10
light_rain_threshold = 50

# Convert rainfall amounts to states
def get_state(rainfall):
    if rainfall <= dry_threshold:
        return 'Dry'
    elif rainfall <= light_rain_threshold:
        return 'Light Rain'
    else:
        return 'Heavy Rain'

# Prepare the historical states from your dataset
historical_states = []
for month in ['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC']:
    monthly_states = df[month].apply(get_state)
    historical_states.extend(monthly_states)

# Define the states
states = ['Dry', 'Light Rain', 'Heavy Rain']

# Initialize a transition matrix
transition_matrix = np.zeros((len(states), len(states)))

# Create a mapping from state names to indices
state_to_index = {state: i for i, state in enumerate(states)}

# Calculate the transition counts
for (current_state, next_state) in zip(historical_states[:-1], historical_states[1:]):
    current_index = state_to_index[current_state]
    next_index = state_to_index[next_state]
    transition_matrix[current_index, next_index] += 1

# Convert counts to probabilities
for i in range(len(states)):
    if transition_matrix[i].sum() != 0:
        transition_matrix[i] /= transition_matrix[i].sum()

print("Transition Matrix:")
print(transition_matrix)

# Initial state probabilities based on historical data
initial_counts = pd.Series(historical_states).value_counts(normalize=True)
initial_probabilities = np.array([initial_counts.get(state, 0) for state in states])

print("Initial Probabilities:")
print(initial_probabilities)

# Function to generate a sequence of states using the Markov Chain model
def generate_sequence(num_steps):
    sequence = []
    current_state = np.random.choice(states, p=initial_probabilities)
    sequence.append(current_state)
    
    for _ in range(num_steps - 1):
        current_state_index = state_to_index[current_state]
        next_state = np.random.choice(states, p=transition_matrix[current_state_index])
        sequence.append(next_state)
        current_state = next_state
    
    return sequence

# Generate a sequence of states (e.g., for the next 12 months)
num_steps = 12
predicted_sequence = generate_sequence(num_steps)
print("Predicted Sequence of States for the next {} months:".format(num_steps))
print(predicted_sequence)
```

The output of your Markov Chain model provides two main components: the transition matrix and the predicted sequence of states for the next 12 months. Here's what each part means:

This matrix describes the probabilities of transitioning from one state to another. Each row represents the current state, and each column represents the next state. The values in the matrix are the probabilities of moving from the current state (row) to the next state (column).

The first row corresponds to the current state "Dry":

The probability of staying in "Dry" is approximately 78.6%.
The probability of transitioning to "Light Rain" is approximately 20.4%.
The probability of transitioning to "Heavy Rain" is approximately 1.0%.
The second row corresponds to the current state "Light Rain":

The probability of transitioning to "Dry" is approximately 16.5%.
The probability of staying in "Light Rain" is approximately 70.8%.
The probability of transitioning to "Heavy Rain" is approximately 12.7%.
The third row corresponds to the current state "Heavy Rain":

The probability of transitioning to "Dry" is approximately 0.4%.

The probability of staying in "Heavy Rain" is approximately 91.6%.

Dynamic downscaling involves using regional climate models (RCMs) to obtain high-resolution climate projections from global climate models (GCMs). The process requires a substantial amount of computational resources and is typically carried out in specialized climate research institutions. However, we can create a simplified example of dynamic downscaling using machine learning techniques.

Here's a general approach to performing dynamic downscaling using machine learning, specifically focusing on rainfall prediction:
1. Prepare the Dataset
Assume you have two datasets:
Global Climate Model (GCM) Outputs: Low-resolution climate data.
Regional Climate Data: High-resolution observed data for training.
2. Preprocessing the Data
Ensure that both datasets have the same variables and time periods for alignment.
3. Train a Machine Learning Model
Use a machine learning model to learn the relationship between the low-resolution GCM outputs and the high-resolution regional data.
4. Downscale the GCM Outputs


```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler

# Load your dataset
# Replace 'your_dataset.csv' with the actual file path or data source
data = pd.read_csv('district_wise_rainfall_normal.csv')

# Display the dataset's columns to ensure they match the index provided
print(data.columns)

# Select the features and target variable
# Assuming you want to predict the 'ANNUAL' rainfall using monthly data
features = ['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC']
target = 'ANNUAL'

# Extract the feature matrix (X) and target vector (y)
X = data[features]
y = data[target]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Standardize the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Train the Random Forest model
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train_scaled, y_train)

# Make predictions
y_pred = model.predict(X_test_scaled)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error (MSE) on Test Data: {mse}')
print(f'Root Mean Squared Error (RMSE) on Test Data: {rmse}')
print(f'R-squared (R) on Test Data: {r2}')

# Example downscaling new data
# Replace with actual new data
new_data = pd.DataFrame({
    'JAN': [20, 30], 'FEB': [25, 35], 'MAR': [15, 25], 'APR': [10, 20],
    'MAY': [5, 15], 'JUN': [50, 60], 'JUL': [60, 70], 'AUG': [70, 80],
    'SEP': [55, 65], 'OCT': [40, 50], 'NOV': [30, 40], 'DEC': [25, 35]
})

new_data_scaled = scaler.transform(new_data)
downscaled_predictions = model.predict(new_data_scaled)
print(f'Downscaled Predictions: {downscaled_predictions}')
```
